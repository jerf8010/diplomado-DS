```python
# Numpy es una biblioteca de Python que proporciona un objeto de matriz multidimensional, varios objetos derivados (como matrices y matrices enmascaradas) 
#y una variedad de rutinas para operaciones rápidas en matrices, que incluyen manipulación matemática, lógica, de formas, clasificación, selección, E/S. , 
#transformadas discretas de Fourier, álgebra lineal básica, operaciones estadísticas básicas, simulación aleatoria y mucho más.


import numpy as np
#Matplotlib es una biblioteca para la generación de gráficos a partir de datos contenidos en listas o arrays 
#en el lenguaje de programación Python y su extensión matemática NumPy
import matplotlib.pyplot as plt
```

```python
#función cuadratica
def my_funcion(x):
  return x**2-5*x+2
def derivada(x):
  return 2*x-5
```

```python
x=np.linspace(-20,20,100) # (inicio, fin, número de muestras a generar)
y=my_funcion(x)
```

```python
plt.plot(x,y)
```

```python
#se tiene que hacer el desenso de gradiente esto es que la derivada de cero
plt.plot(x,y) #función
plt.plot(2.5, my_funcion(2.5),'bo') #bo marcador que representa el punto
```

```python
# se presenta la representación gráfica cuando x=15 y su derivada en ese punto.
init_x=15
init_y=my_funcion(init_x)
dx= derivada (init_x)
px=init_x-dx
py= my_funcion(px)

```

```python
plt.plot(x,y) #pinta la función
plt.plot(init_x,init_y, 'bo') # punto cuando x vale 15
plt.plot(px,py, 'bo') #punto con la derivada cuando x=15
```

```python
#tamaño del gradiente desendiente
#taza de aprendizaje
lr=0.1 #learning rate
init_x=15
init_y=my_funcion(init_x)
dx= derivada (init_x)
px=init_x-dx*lr
py= my_funcion(px)
```

```python
plt.plot(x,y)
plt.plot(init_x,init_y, 'bo') #Punto x=15 
plt.plot(px,py, 'bo')# punto con el gradiente
```

```python
lr=0.1 #learning rate
dx= derivada (init_x)
px=init_x-dx*lr #12.5
py= my_funcion(px)
```

```python
#para hacer el desenso se mete en un bucle
#al hacer el descenso la derivada tiende a cero 

init_x=15
init_y=my_funcion(init_x)
plt.plot(x,y)
plt.plot(init_x,init_y, 'bo')
plt.plot(px,py, 'bo')


for i in range(30): # 10,15,30   con treinta iteraciones se acerca mas a 2.5 el erro es casi cero  2.500046768052395
  px=px - derivada(px) * lr
  py=my_funcion(px)
  plt.plot(px,py, 'bo')
  print(px)

```

```python
#Si se cambia la taza de aprendizaje lr=0.01 los passo seran más pequeños y tardara más en descender 
lr=0.01 #learning rate
init_x=15
init_y=my_funcion(init_x)
dx= derivada (init_x)
px=init_x-dx*lr
py= my_funcion(px)
```

```python
plt.plot(x,y)
plt.plot(init_x,init_y, 'bo')
plt.plot(px,py, 'bo')


for i in range(30): 
  px=px - derivada(px) * lr
  py=my_funcion(px)
  plt.plot(px,py, 'bo')
  print(px)
```

```python
#para el descenso por el otro lado se hace con -15 la pendiente es negativa
lr=0.01 #learning rate
init_x=-15
init_y=my_funcion(init_x)
dx= derivada (init_x)
px=init_x-dx*lr
py= my_funcion(px)
```

```python
plt.plot(x,y)
plt.plot(init_x,init_y, 'bo')
plt.plot(px,py, 'bo')


for i in range(30): # 10,15,30   con treinta iteraciones se acerca mas a 2.5 el erro es casi cero  2.500046768052395
  px=px - derivada(px) * lr
  py=my_funcion(px)
  plt.plot(px,py, 'bo')
  print(px)
```
