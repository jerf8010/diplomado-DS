```python
import pandas as pd
```

```python
cities_df = pd.read_csv('/content/Cities.csv')
```

```python
cities_df.head(5)
```

```python
temperature_normalized = cities_df['temperature'].apply(lambda x: (x - cities_df['temperature'].min()) / (cities_df['temperature'].max() - cities_df['temperature'].min()))
```

```python
temperature_normalized
```

```python
temperature_std = cities_df['temperature'].apply(lambda x: (x - cities_df['temperature'].mean()) / cities_df['temperature'].std())
```

```python
temperature_std
```

```python
country_temperature = dict(cities_df.groupby('country')['temperature'].mean())
```

```python
cities_df['country_temp'] = cities_df.apply(lambda x: country_temperature[x['country']], axis = 1)
```

```python
cities_df.head(5)
```

```python
cities_df['vivible'] = cities_df['temperature'].apply(lambda x: x > 10)
```
