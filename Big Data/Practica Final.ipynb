### Realizar un agrupamiento

```python
print('Agrupamiento: ')
df.groupBy('month').mean().sort('month', ascending=[0]).toPandas()['avg(duration)']
```

### Mostrar la filas ordenas por un campo

```python
print('Filas ordenadas por campo mile: ')
df.orderBy('mile', ascending = [0]).toPandas()
```

### Generar una consulta SQL desde el DataFrame

```python
print('Consulta SQL')

df.select('origin', 'duration') \
    .where(df.month == 1) \
    .groupBy('origin').mean() \
    .orderBy('avg(duration)', ascending = 0).toPandas()
```

### Generar un agrupamiento que muestre funciones de agregacion, minimo tres (sum, max,min, avg)

```python
from pyspark.sql.functions import col, count, avg, max
```

```python
print('Agrupamiento con funciones de agregación')


df.groupBy('month', 'origin') \
    .agg(sum('mile').alias("sum_miles"), \
         avg('duration').alias("avg_duration"), \
         count('origin').alias('total_flights'), \
         max('duration').alias("max_duration") \
     ) \
    .filter(col('avg_duration') > 120)  \
    .sort('max_duration') \
    .show(truncate=False)
```

### Guardar el resultado en una tabla de hive

```python
df.write.partitionBy('duration').mode("overwrite").saveAsTable('table_hive') 
```
