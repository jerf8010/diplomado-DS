## Analizando Moby Dick

```python
from nltk.book import *
text1
```

```python
with open('moby.txt', 'r') as f:
    moby_raw = f.read()
```

### Ejemplo 1

¿Cuántos tokens (palabras y símbolos de puntuación) hay en texto1?

```python
def ejemplo1():
    
    return len(nltk.word_tokenize(moby_raw)), len(text1), len(text1.tokens)

ejemplo1()
```

### Ejemplo 2

¿Cuántos tokens únicos (palabras únicas y símbolos de puntuación) tiene texto1?

```python
def ejemplo2():
    
    return len(set(nltk.word_tokenize(moby_raw))), len(set(text1))

ejemplo2()
```

### Etiquetado de palabras

```python
tag = nltk.pos_tag(text1)
tag
```

### Ejemplo 3

¿Después de lematizar los verbos, cuántos tokens únicos tiene texto1?

```python
from nltk.stem import WordNetLemmatizer

def example_three():
    
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(t[0]) for t in tag if t[1].startswith('VB')]
    
    return len(set(lemmatized))

example_three()
```

### Pregunta 1

¿Cuál es la diversidad léxica del texto de entrada? (ratio entre tokens únicos y el número total de tokens)

```python
def respuesta1():
    respuesta =  len(set(text1)) / len(text1)
    return respuesta

respuesta1()
```

### Pregunta 2

¿Qué porcentaje de tokens es 'whale'or 'Whale'?

```python
def respuesta2():
    respuesta = 100 * (text1.count('whale')+text1.count('Whale')) / len(text1)
    return respuesta

respuesta2()
```

### Pregunta 3
¿Cuáles son los 20 tokens más frecuentes (únicos) en el texto? ¿Cuál es su frecuencia?

*Esta función debería devolver una lista de 20 tuplas donde cada tupla tiene la forma `(token, frecuencia)`. La lista debe ordenarse en orden descendente de frecuencia.*

```python
from collections import Counter
def respuesta3():
    respuesta = Counter()
    for word in text1.tokens:
        respuesta[word] += 1
    return respuesta.most_common(20)

# Counter(moby_tokens).most_common(20)
''' OTRA SOLUCIÓN <tarda mucho>
    respuesta = []
    for w in text1.tokens:
        respuesta.append((w, text1.count(w)))
    respuesta = sorted(respuesta, key=lambda rpta: rpta[1])
    return respuesta[0:21]
'''
respuesta3()
```

### Pregunta 4

¿Qué tokens tienen una longitud superior a 5 y frecuencia de más de 150?

*Esta función debería devolver una lista ordenada alfabeticamente de los tokens que coinciden con las restricciones anteriores. Para ordenar la lista, usar `sorted ()`*

```python
def respuesta4():
    fdist = FreqDist(text1)
    respuesta = sorted([word for word in set(text1.tokens) if len(word) > 5 and fdist[word] > 150])
    return respuesta

''' OTRA SOLUCIÓN <tarda un poco, pero da el mismo resultado>  
    respuesta = []
    for word in set(texto1.tokens):
        if len(word) > 5 and texto1.count(word) > 150:
            respuesta.append(word)
'''

respuesta4()
```

### Pregunta 5

Encontrar la palabra mas larga en texto1 y la longitud de dicha palabra.

*Esta función debe devolver una tupla `(palabra, longitud)`.*

```python
def respuesta5():
    words_len = [(word, len(word)) for word in set(text1.tokens)]
    respuesta = sorted(words_len, key=lambda word_len: word_len[1], reverse=True)
    return respuesta[0]

respuesta5()
```

### Pregunta 6

Qué palabras únicas tienen una frecuencia de más de 2000? ¿Cuál es su frecuencia?

"Sugerencia: es conveniente utilizar` isalpha () `para comprobar si el token es una palabra y no una puntuación".

* Esta función debe devolver una lista de tuplas de la forma `(frecuencia, palabra)` ordenadas en orden descendente de frecuencia. *

```python
def respuesta6():
    fdist = FreqDist(text1)
    words_unique = [(fdist[word], word) for word in set(text1.tokens) if word.isalpha() and fdist[word] > 2000]
    respuesta = sorted(words_unique, key=lambda word: word[0], reverse=True)
    return respuesta

respuesta6()
```

### Pregunta 7
¿Cuál es el número promedio de tokens por oración?

```python
import statistics
def respuesta7():
    sentences = nltk.sent_tokenize(moby_raw)
    respuesta = [len(nltk.word_tokenize(w)) for w in sentences]
    return statistics.mean(respuesta)

respuesta7()
```

### Pregunta 8

¿Cuáles son las 5 etiquetas gramaticales más frecuentes en el texto? ¿Cuál es su frecuencia?

* Esta función debe devolver una lista de tuplas de la forma `(etiqueta_gramatical, frecuencia)` ordenadas en orden descendente de frecuencia. *

```python
def respuesta8():
    tag_fd = nltk.FreqDist(tg for (word, tg) in tag)
    respuesta = tag_fd.most_common(5)
    return respuesta
    
respuesta8()
```
